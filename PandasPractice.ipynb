```python
import pandas as pd
```

```python
df = pd.read_csv('traffic_accidents.csv')
df
```

```python
#Replace null with median , mean or mode
#df.dropna(inplace = True)
x = df['injuries_total'].median()

df['injuries_total'].fillna(value = x, inplace = True)
df
```

```python
#Removing Duplicates
df.drop_duplicates(inplace = True)
df
```
